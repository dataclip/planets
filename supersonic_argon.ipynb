```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 60, dtype = np.float)
M = (3.26*(0.1*x)**0.67) - (0.5*(0.67/2.67))/(3.26*(0.1*x)**0.67)
T = 300*(1+(0.67/2)*M**2)**(-1)
C = (0.67*8.314*300/.039)**0.5 # C is a constant to calculate velocity, V
V = M*C*(1+(0.67/2)*M**2)**(-0.5)
n = (1+(0.67/2)*M**2)**(-1/0.67)

plt.subplot(2, 2, 1)
plt.plot(x, M, "bo-")
plt.xlabel('x/mm')
plt.ylabel('M')
plt.text(20, 10, 'Axisymmetric Expansion')


plt.subplot(2, 2, 2)
plt.plot (x, T, 'bo-')
plt.xlabel('x/mm')
plt.ylabel('T/K')
plt.show()

plt.subplot(2, 2, 3)
plt.plot(x, V, "bo-")
plt.xlabel('x/mm')
plt.ylabel('V(m/s)')
plt.show()

plt.subplot(2, 2, 4)
plt.plot(x, n, "bo-")
plt.xlabel('x/mm')
plt.ylabel('n/n0')
plt.show()

plt.show()
```
